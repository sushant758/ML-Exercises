```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df['Species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Species',axis=1)
y = df['Species']
```

```python
# when you will use cross validation , you dont
# have to use train test split

# But here, now  i am using train test split
```

```python
X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size=0.3 , random_state=42)
```

```python
# using KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors = 13)
```

```python
model_knn.fit(X_train,y_train)
```

```python
model_knn.score(X_test,y_test)
```

```python
# now it means that model is getting overfit
```

# SVM

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(gamma='auto')
```

```python
model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test)
```

# Now lets use grid search CV

```python
# this can reduce the overfitting of our model
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# now perform hyperparameter tunning / create grids
```

```python
classifier = GridSearchCV((model_svm),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False)
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.cv_results_
```
